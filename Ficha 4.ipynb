10. Escreva um programa Python para criar um gráfico de barras de pontuações por grupo e género.

Dados da amostra:
Médias (homens) = (22, 30, 35, 35, 26)

15. Escreva um programa Python para criar um gráfico de barras horizontais com cores ordenadas 
de forma diferente. 

```python
areas = [ 
    ['Linguas','Ciencia','Matematica'], 
    ['Ciencia', 'Matematica','Linguas'], 
    ['Matematica','Linguas','Ciencia']
]

notas = [
    {'Linguas':75, 'Ciencia':88, 'Matematica':96},
    {'Linguas':71, 'Ciencia':95, 'Matematica':92},
    {'Linguas':75, 'Ciencia':90, 'Matematica':89}
]
```

```python

labels = np.arange(len(areas))
alunos = ["Aluno"+str(i) for i in range(1, len(labels)+1)]
cores = {"Linguas": "g", "Ciencia": "r", "Matematica": "b"}
barras = dict()

for area, nota, i in zip(areas, notas, labels):

    last_n = 0
    for a in area:
        b = plt.barh(i, nota[a], left=last_n, color=cores[a], label=a)

        # Guardar as barras com label diferente, para poder fazer a legenda depois
        if not (b.get_label() in barras):
            barras[a] = b
            
        last_n += nota[a]

plt.minorticks_on()
plt.grid(True, color="gray")
plt.grid(True, which="minor")
plt.yticks(labels, alunos)
plt.legend(barras)
```

16. Escreva um programa Python para criar um gráfico de barras empilhadas e adicione o rótulo a 
cada secção (percentagem aleatória entre 5% e 20%).

```python
pessoas = ('G1','G2','G3','G4','G5','G6','G7','G8')
segmentos = 4
# Dados multidimensionais
dados = [
    [3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.4473477], 
    [11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
    [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
    [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]
]
```

```python
labels = np.arange(len(pessoas))
cols =  range(len(dados[0]))
cores = ["r", "g", "white", "pink"]

sums = [sum(dados[i]) for i in range(segmentos)]

for col in cols:
    last_n = 0
    
    for seg in range(segmentos):
        barra = plt.barh(col, dados[seg][col], left=last_n, linewidth=1, edgecolor="black", color=cores[seg])          
        last_n += dados[seg][col]

        w = last_n
        plt.annotate('%.0f%%' % (dados[seg][col]/sums[seg] * 100),
          xy=(w, col),
          xytext=(0, 4),
          textcoords="offset points",
          ha='right', va='center')


plt.yticks(labels, pessoas)
plt.xlabel("Pontuações")
```
